```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd 
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
data = requests.get(url).text 
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
tables = soup.find_all('table')
```

```python
len(tables)
```

```python
for index, table in enumerate(tables):
  if("10 most densely populated countries" in str(table)): 
    table_index = index 
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
population_data = pd.DataFrame(columns=["Rank", "Country","Population", "Area","Density"])

for row in tables[table_index].tbody.find_all('tr'): 
  col = row.find_all('td')
  if(col !=[]): 
    rank = col[0].text 
    country = col[1].text 
    population = col[2].text.strip()
    area = col[3].text.strip()
    density = col[4].text.strip()

    population_data = population_data.append({"Rank":rank, "Country":country, "Population":population,'Area':area, "Density":density}, ignore_index=True)
  
population_data
```

```python
url_srx = "https://www.srx.com.sg/new-condo-launch"
```

```python
data = requests.get(url_srx).text 
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
tables = soup.find_all('table')
```

```python
len(tables)
```

```python
for index, table in enumerate(tables):
  if("Upcoming Launches" in str(table)): 
    table_index = index 
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
new_launch_data = pd.DataFrame(columns=['Project Name', 'Location(District',
                                        'Launch Date', 'Top', 'Total Units'])

for row in tables[table_index].tbody.find_all('tr'): 
  col = row.find_all('td')
  if(col !=[]): 
    project_name = col[0].text.strip()
    location = col[1].text 
    launch_date = col[2].text.strip()
    top = col[3].text.strip()
    total_units = col[4].text 

    new_launch_data = new_launch_data.append({'Project Name': project_name, 'Location(District':location,
                                             'Launch Date': launch_date, 'Top':top, 'Total Units': total_units}, ignore_index=True)
new_launch_data
```
